````python
# CELDA INTERACTIVA 2: ANALOG√çAS ESTRAT√âGICAS
print("üéÆ MAPA DE ANALOG√çAS: Ajedrez ‚Üí FinTech")
print("=" * 50)

class EstrategiasCompetitivas:
    def __init__(self):
        self.analogias = {
            "Apertura": "Posicionamiento inicial en el mercado",
            "Desarrollo de piezas": "Optimizaci√≥n de recursos y herramientas",
            "Control del centro": "Dominio del mercado principal",
            "Ataque al rey": "Captura de clientes premium",
            "Defensa flexible": "Protecci√≥n de cartera existente",
            "Estrategia a largo plazo": "Planificaci√≥n de crecimiento sostenible"
        }

    def mostrar_analogias(self):
        print("‚ôüÔ∏è  AJEDREZ\t\t\tüíº FINTECH")
        print("-" * 55)
        for ajedrez, fintech in self.analogias.items():
            print(f"‚Ä¢ {ajedrez:<20} ‚Üí {fintech}")

    def calcular_ventaja_estrategica(self, profundidad_analisis):
        """Calcula la ventaja de pensar m√∫ltiples pasos adelante"""
        ventaja_base = 100  # puntos base
        ventaja = ventaja_base * (1.5 ** profundidad_analisis)
        return ventaja

# Demostraci√≥n
estrategia = EstrategiasCompetitivas()
estrategia.mostrar_analogias()

print(f"\nüìà VENTAJA ESTRAT√âGICA POR PROFUNDIDAD:")
for profundidad in [1, 2, 3, 4]:
    ventaja = estrategia.calcular_ventaja_estrategica(profundidad)
    print(f"   Profundidad {profundidad}: +{ventaja:.0f} puntos de ventaja")

# %% [markdown]
# ### üí° **Insight Clave:**
# Pensar **3-4 movimientos adelante** (profundidad de an√°lisis) multiplica nuestra ventaja competitiva exponencialmente, similar a c√≥mo los grandes maestros de ajedrez dominan el juego.
#

# %% [markdown]
# # ‚ö° **SEGMENTO 2: Algoritmo Minimax - N√∫cleo de la Decisi√≥n Adversarial**
#

# %%
````

````python
# %% [markdown]
# # ‚ö° **SEGMENTO 2: Algoritmo Minimax - N√∫cleo de la Decisi√≥n Adversarial**
#

# %%
# CELDA INTERACTIVA 3: IMPLEMENTACI√ìN MINIMAX VISUAL - CORREGIDA
print("‚ö° ALGORITMO MINIMAX - TOMA DE DECISIONES ADVERSARIAL")
print("=" * 65)

import time
import matplotlib.pyplot as plt
import numpy as np

class MinimaxFinFlow:
    def __init__(self):
        self.historial_decisiones = []

    def calcular_utilidad(self, tasa, riesgo_cliente, contexto_mercado):
        """Calcula utilidad considerando m√∫ltiples factores"""
        utilidad_base = 1000

        # Factores de ajuste
        factor_riesgo = 0.8 if riesgo_cliente == "alto" else 1.2
        factor_competencia = 0.9 if contexto_mercado == "competitivo" else 1.1
        factor_tasa = 1.0 - (abs(tasa - 10) * 0.05)  # Penaliza desviaci√≥n de tasa √≥ptima

        utilidad = utilidad_base * factor_riesgo * factor_competencia * factor_tasa
        return utilidad

    def minimax(self, profundidad, es_maximizador, tasa_actual, riesgo, contexto):
        if profundidad == 0:
            utilidad = self.calcular_utilidad(tasa_actual, riesgo, contexto)
            self.historial_decisiones.append(("hoja", tasa_actual, utilidad))
            return utilidad, tasa_actual

        if es_maximizador:  # FinFlow
            print(f"üî∫ Nivel {profundidad}: FinFlow eval√∫a tasas...")
            mejor_utilidad = -float('inf')
            mejor_tasa = None

            for tasa in [8, 9, 10, 11, 12]:
                utilidad, _ = self.minimax(profundidad-1, False, tasa, riesgo, contexto)

                if utilidad > mejor_utilidad:
                    mejor_utilidad = utilidad
                    mejor_tasa = tasa
                    print(f"   ‚úÖ Tasa {tasa}% ‚Üí Utilidad: {utilidad:.1f} *NUEVA MEJOR*")
                else:
                    print(f"   ‚ùå Tasa {tasa}% ‚Üí Utilidad: {utilidad:.1f}")

            self.historial_decisiones.append(("max", mejor_tasa, mejor_utilidad))
            return mejor_utilidad, mejor_tasa

        else:  # Competidor
            print(f"üîª Nivel {profundidad}: Competidor responde...")
            peor_utilidad = float('inf')

            for tasa_competidor in [7, 8, 9, 10, 11]:
                utilidad, _ = self.minimax(profundidad-1, True, tasa_competidor, riesgo, contexto)

                if utilidad < peor_utilidad:
                    peor_utilidad = utilidad
                    print(f"   üéØ Competidor elige tasa {tasa_competidor}% (Utilidad para nosotros: {utilidad:.1f})")

            return peor_utilidad, None

# Ejecutar Minimax - CORREGIDO (sin typo)
print("üß™ SIMULACI√ìN: Cliente riesgo medio, mercado competitivo")
minimax_engine = MinimaxFinFlow()
utilidad_optima, tasa_optima = minimax_engine.minimax(2, True, 10, "medio", "competitivo")  # CORREGIDO: minimax_engine

print(f"\nüéØ DECISI√ìN FINAL MINIMAX:")
print(f"   üí∞ Tasa recomendada: {tasa_optima}%")
print(f"   üìà Utilidad esperada: ${utilidad_optima:.2f}")
print(f"   üõ°Ô∏è  Garantizada incluso en PEOR escenario competitivo")

# Visualizaci√≥n adicional del √°rbol de decisiones
print(f"\nüìä HISTORIAL DE DECISIONES:")
for i, (tipo, tasa, utilidad) in enumerate(minimax_engine.historial_decisiones[-5:]):  # √öltimas 5 decisiones
    print(f"   {i+1}. {tipo.upper()} - Tasa: {tasa if tasa else 'N/A'}% - Utilidad: {utilidad:.1f}")

# %% [markdown]
# ### üß† **Explicaci√≥n Conceptual:**
# **Minimax** asegura que nuestra decisi√≥n sea **robusta** - maximiza nuestra ganancia m√≠nima garantizada, asumiendo que el competidor siempre elegir√° lo que m√°s nos perjudique.
#
# **En FinFlow:** Esto significa que elegimos la tasa que nos da el mejor resultado posible, incluso si los competidores responden de la manera m√°s agresiva contra nosotros.
#

# %%
# CELDA EXTRA: VISUALIZACI√ìN DEL √ÅRBOL MINIMAX
print("\nüå≥ VISUALIZACI√ìN DEL √ÅRBOL DE DECISIONES MINIMAX")
print("=" * 55)

def visualizar_arbol_minimax():
    # Crear datos de ejemplo para visualizaci√≥n
    niveles = ['Ra√≠z\n(FinFlow)', 'Nivel 1\n(Competidor)', 'Nivel 2\n(FinFlow)', 'Hojas\n(Resultado)']
    nodos_por_nivel = [1, 3, 9, 27]  # Crecimiento exponencial t√≠pico

    plt.figure(figsize=(10, 6))
    plt.plot(niveles, nodos_por_nivel, marker='o', linewidth=3, markersize=8, color='#2E86AB')
    plt.fill_between(niveles, nodos_por_nivel, alpha=0.2, color='#2E86AB')
    plt.title('üå≥ Crecimiento Exponencial del √Årbol de Decisiones', fontsize=14, fontweight='bold')
    plt.ylabel('N√∫mero de Nodos', fontweight='bold')
    plt.xlabel('Profundidad del √Årbol', fontweight='bold')
    plt.grid(True, alpha=0.3)

    # Anotar el crecimiento
    for i, (nivel, nodos) in enumerate(zip(niveles, nodos_por_nivel)):
        plt.annotate(f'{nodos} nodos', (nivel, nodos), textcoords="offset points",
                    xytext=(0,10), ha='center', fontweight='bold')

    plt.tight_layout()
    plt.show()

# Mostrar explicaci√≥n del crecimiento
print("üìà EXPLICACI√ìN DEL CRECIMIENTO EXPONENCIAL:")
print("‚Ä¢ Ra√≠z: 1 decisi√≥n inicial de FinFlow")
print("‚Ä¢ Nivel 1: 3 posibles respuestas del competidor por cada decisi√≥n nuestra")
print("‚Ä¢ Nivel 2: 9 posibles contra-respuestas nuestras")
print("‚Ä¢ Hojas: 27 resultados finales posibles")
print("\nüí° Esto demuestra por qu√© necesitamos optimizaciones como la poda alfa-beta!")

# Ejecutar visualizaci√≥n
visualizar_arbol_minimax()
````

````python
# CELDA INTERACTIVA 4: PODA ALFA-BETA CON M√âTRICAS
print("üöÄ PODA ALFA-BETA - OPTIMIZANDO LA B√öSQUEDA ADVERSARIAL")
print("=" * 70)

class PodaAlfaBeta:
    def __init__(self):
        self.nodos_evaluados = 0
        self.nodos_podados = 0

    def poda_alfa_beta(self, profundidad, es_maximizador, alfa, beta, tasa_actual, riesgo, contexto, nodo_id=""):
        self.nodos_evaluados += 1

        if profundidad == 0:
            utilidad = self.calcular_utilidad(tasa_actual, riesgo, contexto)
            return utilidad, tasa_actual

        if es_maximizador:
            print(f"üî∫ MAX [{nodo_id}] Alfa:{alfa:.1f} Beta:{beta:.1f}")
            mejor_utilidad = -float('inf')
            mejor_tasa = None

            for i, tasa in enumerate([8, 9, 10, 11, 12]):
                utilidad, _ = self.poda_alfa_beta(profundidad-1, False, alfa, beta, tasa, riesgo, contexto, f"T{i+1}")

                if utilidad > mejor_utilidad:
                    mejor_utilidad = utilidad
                    mejor_tasa = tasa
                    print(f"   ‚úÖ Tasa {tasa}% ‚Üí Utilidad: {utilidad:.1f}")

                alfa = max(alfa, mejor_utilidad)

                # CONDICI√ìN DE PODA
                if beta <= alfa:
                    self.nodos_podados += (5 - (i + 1))  # Nodos restantes en este nivel
                    print(f"   ‚úÇÔ∏è  PODA! Saltando {5 - (i + 1)} nodos (Beta {beta:.1f} <= Alfa {alfa:.1f})")
                    break

            return mejor_utilidad, mejor_tasa

        else:
            print(f"üîª MIN [{nodo_id}] Alfa:{alfa:.1f} Beta:{beta:.1f}")
            peor_utilidad = float('inf')

            for i, tasa_comp in enumerate([7, 8, 9, 10, 11]):
                utilidad, _ = self.poda_alfa_beta(profundidad-1, True, alfa, beta, tasa_comp, riesgo, contexto, f"C{i+1}")

                if utilidad < peor_utilidad:
                    peor_utilidad = utilidad
                    print(f"   üéØ Competidor elige tasa {tasa_comp}%")

                beta = min(beta, peor_utilidad)

                # CONDICI√ìN DE PODA
                if beta <= alfa:
                    self.nodos_podados += (5 - (i + 1))
                    print(f"   ‚úÇÔ∏è  PODA! Saltando {5 - (i + 1)} respuestas competidor")
                    break

            return peor_utilidad, None

    def calcular_utilidad(self, tasa, riesgo, contexto):
        # Implementaci√≥n simplificada
        return 1000 - abs(tasa - 10) * 50

# Comparar eficiencia
print("üìä COMPARACI√ìN: MINIMAX vs PODA ALFA-BETA")
print("-" * 45)

# Minimax tradicional
minimax_simple = MinimaxFinFlow()
inicio_minimax = time.time()
util_minimax, tasa_minimax = minimax_simple.minimax(3, True, 10, "medio", "competitivo")
tiempo_minimax = time.time() - inicio_minimax

# Poda Alfa-Beta
poda_engine = PodaAlfaBeta()
inicio_poda = time.time()
util_poda, tasa_poda = poda_engine.poda_alfa_beta(3, True, -float('inf'), float('inf'), 10, "medio", "competitivo", "RAIZ")
tiempo_poda = time.time() - inicio_poda

print(f"\nüéØ RESULTADOS DE EFICIENCIA:")
print(f"‚Ä¢ Minimax tradicional: {tiempo_minimax:.4f}s - {minimax_simple.historial_decisiones.__len__()} nodos")
print(f"‚Ä¢ Poda Alfa-Beta: {tiempo_poda:.4f}s - {poda_engine.nodos_evaluados} nodos evaluados")
print(f"‚Ä¢ Nodos podados: {poda_engine.nodos_podados}")
print(f"‚Ä¢ Mejora de velocidad: {((tiempo_minimax-tiempo_poda)/tiempo_minimax)*100:.1f}%")

# %% [markdown]
# ### üí° **Ventaja de la Poda Alfa-Beta:**
# Elimina **ramas completas** del √°rbol de decisiones que no afectar√°n el resultado final, permitiendo:
# - ‚úÖ **An√°lisis m√°s profundo** en el mismo tiempo
# - ‚úÖ **Decisiones en tiempo real**
# - ‚úÖ **Menor consumo computacional**
#

# %% [markdown]
# # üß† **SEGMENTO 4: Backpropagation como Soluci√≥n Estrat√©gica**
#

# %%
````

````python
# CELDA INTERACTIVA 5: BACKPROPAGATION EN DECISIONES ESTRAT√âGICAS
print("üß† BACKPROPAGATION ESTRAT√âGICO - APRENDIENDO DE RESULTADOS")
print("=" * 75)

class BackpropagationEstrategico:
    def __init__(self):
        self.pesos_decisiones = {
            'factor_riesgo': 0.8,
            'factor_competencia': 1.0,
            'factor_tasa_optima': 1.2
        }
        self.tasa_aprendizaje = 0.1
        self.historial_errores = []

    def forward_pass(self, tasa_decisi√≥n, riesgo, contexto):
        """Paso forward - calcular predicci√≥n de utilidad"""
        utilidad_predicha = (
            self.pesos_decisiones['factor_riesgo'] * (0.8 if riesgo == "alto" else 1.2) +
            self.pesos_decisiones['factor_competencia'] * (0.9 if contexto == "competitivo" else 1.1) +
            self.pesos_decisiones['factor_tasa_optima'] * (1.0 - abs(tasa_decisi√≥n - 10) * 0.05)
        ) * 500  # Escalar

        return utilidad_predicha

    def backward_pass(self, tasa_decisi√≥n, utilidad_real, riesgo, contexto):
        """Paso backward - ajustar pesos basado en error"""
        utilidad_predicha = self.forward_pass(tasa_decisi√≥n, riesgo, contexto)
        error = utilidad_real - utilidad_predicha
        self.historial_errores.append(abs(error))

        print(f"üìä BACKPROPAGATION ITERACI√ìN:")
        print(f"   Predicci√≥n: {utilidad_predicha:.1f}, Real: {utilidad_real:.1f}")
        print(f"   Error: {error:.1f}")

        # Ajustar pesos (simplificado)
        ajuste = error * self.tasa_aprendizaje
        for clave in self.pesos_decisiones:
            self.pesos_decisiones[clave] += ajuste * 0.3

        print(f"   Nuevos pesos: {self.pesos_decisiones}")
        return error

    def entrenar_con_historial(self, historial_decisiones):
        """Entrenar con datos hist√≥ricos"""
        print("\nüéØ ENTRENAMIENTO CON DATOS HIST√ìRICOS")
        print("-" * 40)

        for i, decision in enumerate(historial_decisiones[:5]):  # Primeras 5 para ejemplo
            tasa, util_real, riesgo, contexto = decision
            error = self.backward_pass(tasa, util_real, riesgo, contexto)
            print(f"   √âpoca {i+1}: Error = {error:.2f}")

        print(f"\nüìà EVOLUCI√ìN DEL ERROR: {self.historial_errores}")

# Simular historial de decisiones
historial_ejemplo = [
    (8, 850, "medio", "competitivo"),
    (10, 920, "bajo", "estable"),
    (12, 780, "alto", "competitivo"),
    (9, 890, "medio", "estable"),
    (11, 810, "alto", "competitivo")
]

# Entrenar modelo
bp_estrategico = BackpropagationEstrategico()
bp_estrategico.entrenar_con_historial(historial_ejemplo)

print(f"\nüí° APRENDIZAJE ESTRAT√âGICO COMPLETADO:")
print(f"   El sistema ahora hace mejores predicciones de utilidad")
print(f"   Aprende de decisiones pasadas para mejorar futuras decisiones")

# %% [markdown]
# ### üîÑ **Backpropagation en Contexto Estrat√©gico:**
#
# **¬øQu√© es?** Un mecanismo de aprendizaje que ajusta nuestros modelos bas√°ndose en el error entre predicciones y resultados reales.
#
# **Aplicaci√≥n en FinFlow:**
# - ‚úÖ **Aprende** de cada decisi√≥n de tasas tomada
# - ‚úÖ **Ajusta** factores de riesgo y competencia
# - ‚úÖ **Mejora** predicciones de utilidad futuras
# - ‚úÖ **Adapta** la estrategia al mercado cambiante
#

# %% [markdown]
# # üí° **SEGMENTO 5: Sistema H√≠brido Propuesto**
#

# %%
````

````python
# CELDA INTERACTIVA 6: ARQUITECTURA DEL SISTEMA H√çBRIDO
print("üí° ARQUITECTURA SISTEMA H√çBRIDO FINFLOW IA")
print("=" * 60)

class SistemaHibridoFinFlow:
    def __init__(self):
        self.modulos = {
            'minimax': 'Decisi√≥n adversarial inmediata',
            'poda_alfa_beta': 'Optimizaci√≥n de b√∫squeda',
            'backpropagation': 'Aprendizaje continuo',
            'monte_carlo': 'Simulaci√≥n de escenarios',
            'refuerzo': 'Estrategia a largo plazo'
        }

    def mostrar_arquitectura(self):
        print("üèóÔ∏è  ARQUITECTURA PROPUESTA:")
        print("‚îå‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îê")
        for modulo, descripcion in self.modulos.items():
            print(f"‚îÇ üéØ {modulo:<15} ‚Üí {descripcion:<25} ‚îÇ")
        print("‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò")

    def flujo_decision(self, cliente, contexto):
        print(f"\nüîÑ FLUJO DE DECISI√ìN PARA CLIENTE: {cliente}")
        print("=" * 45)

        pasos = [
            "1. üìä An√°lisis perfil riesgo y historial",
            "2. üéØ Minimax con poda para decisi√≥n base",
            "3. üß† Backpropagation para ajuste fino",
            "4. üìà Simulaci√≥n Monte Carlo de escenarios",
            "5. üöÄ Decisi√≥n final con aprendizaje por refuerzo"
        ]

        for paso in pasos:
            print(f"   {paso}")
            time.sleep(0.5)  # Efecto visual

        return "Tasa optimizada + L√≠mite cr√©dito + Estrategia competencia"

    def calcular_beneficios_esperados(self):
        beneficios = {
            'Velocidad decisi√≥n': '+65%',
            'Precisi√≥n predicci√≥n': '+40%',
            'Reducci√≥n errores': '-55%',
            'Mejora m√°rgenes': '+12%',
            'Captura clientes': '+25%'
        }

        print("\nüìä BENEFICIOS ESPERADOS DEL SISTEMA H√çBRIDO:")
        for beneficio, valor in beneficios.items():
            print(f"   ‚úÖ {beneficio}: {valor}")

# Demostraci√≥n
sistema_hibrido = SistemaHibridoFinFlow()
sistema_hibrido.mostrar_arquitectura()
sistema_hibrido.flujo_decision("Cliente Premium XYZ", "Mercado Competitivo")
sistema_hibrido.calcular_beneficios_esperados()

# %% [markdown]
# # ü§î **PREGUNTAS PARA EL EQUIPO**
#
# ## üí¨ **Discusi√≥n Colaborativa:**
#
# **1. ¬øQu√© otros algoritmos adversariales podr√≠amos integrar?**

# ```python
# # Ejemplo: Algoritmo Expectiminimax para incertidumbre
# algoritmos_candidatos = ["Expectiminimax", "Monte Carlo Tree Search", "Algoritmo negamax"]
# ```
#
# **2. ¬øC√≥mo validar√≠amos la efectividad en diferentes segmentos de clientes?**
#
# **3. ¬øQu√© m√©tricas espec√≠ficas deber√≠amos trackear para medir el ROI?**
#
# **4. ¬øC√≥mo manejar competidores que no siguen comportamiento racional?**
#
# **5. ¬øQu√© aspectos √©ticos considerar en la implementaci√≥n?**
#

# %% [markdown]
# <div align="center">
#
# # üéØ **CONCLUSI√ìN PRINCIPAL**
#
# **Los algoritmos de b√∫squeda adversarial, combinados con backpropagation estrat√©gico, pueden transformar a FinFlow de reactiva a proactiva en el competitivo mercado fintech.**
#
# *¬°Quedo atenta a sus valiosos aportes!* üöÄüí°
#
# </div>
````

# 1. ¬øQu√© otros algoritmos adversariales podr√≠amos integrar?
